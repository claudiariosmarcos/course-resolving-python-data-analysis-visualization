## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=footballteams,y=championsleagues)
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=footballteams,y=championsleagues,s=5)
```

### Change the color

```python
plt.scatter(x=footballteams,y=championsleagues,c='red')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=footballteams,y=championsleagues,marker=2)
```

### Change the border
