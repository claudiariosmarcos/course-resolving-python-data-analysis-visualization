# Data Visualizations with Excel Data Tables

## Load Excel in Python

- gapminder.xlsx

```python
import pandas
df_countries = pandas.read_excel('gapminder.xlsx')
```

```python
df_countries
```

## Create a scatter plot

### Import matplotlib library

```python
import matplotlib.pyplot as plt
```

### Access column data separately

```python
df_countries.money_per_person
```

### Pass data to scatter function

```python
plt.scatter(x = df_countries.life_expectancy,y = df_countries.money_per_person)
```

## Customize the plot

### Change the size

```python
plt.scatter(x = df_countries.life_expectancy,y = df_countries.money_per_person, s=5)
```

### Change the color

```python
plt.scatter(x = df_countries.life_expectancy,y = df_countries.money_per_person,c='purple')
```

### Change the marker

```python
plt.scatter(x = df_countries.life_expectancy,y = df_countries.money_per_person,marker='x')
```

### Change the border

```python
plt.scatter(x = df_countries.life_expectancy,y = df_countries.money_per_person,edgecolors='red')
```
